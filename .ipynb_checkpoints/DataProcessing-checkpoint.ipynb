```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel("Data/2017_PM10_1g.xlsx") # load
```

```python
krk_index = []
krk_names = []
for index, station in enumerate(df.iloc[0,:]):
    if "krak" in station.lower():
        print(index, station)
        krk_index.append(index)
        krk_names.append(station)
```

```python
krk_df = df.iloc[5:,krk_index]
krk_df.columns = krk_names
krk_df.to_csv("Data/2017_PM10_Krk.csv", index=False)
krk_df.head()
```

# Select best stations

```python
count_nans = krk_df.isna().sum()
best_stations_index = count_nans.nsmallest(3).index
best_stations = krk_df[best_stations_index]
best_stations.head()
```

# Preprocessing

```python
def preprocessData(data):
    data_dot = data.str.replace(",",".")
    data_numeric = pd.to_numeric(data_dot)
```

```python
best_stations_df = pd.DataFrame([pd.to_numeric(best_stations[station].str.replace(",",".")) for station in best_stations])
best_stations_df = best_stations_df.transpose()
best_stations_df.head()
```

# Plotting

```python
for station in best_stations_df:
    best_stations_df[station].plot(figsize=(12,8))
```

```python
best_stations_df.isna().sum()
```

# Handling missing values
I take the simplest approach and take mean of two neighbours

```python
for i in best_stations_df['MpKrakAlKras'][best_stations_df['MpKrakAlKras'].isna()].index:
    if not(np.isnan(best_stations_df['MpKrakAlKras'].loc[i-1]) or np.isnan(best_stations_df['MpKrakAlKras'].loc[i+1])):
        best_stations_df['MpKrakAlKras'].loc[i] = (best_stations_df['MpKrakAlKras'].loc[i-1] + best_stations_df['MpKrakAlKras'].loc[i+1])/2
```

```python
sum(best_stations_df['MpKrakAlKras'].isna())
```

# Moving average vs normal data
moving average also handles missing values

```python
avg = best_stations_df['MpKrakAlKras'].rolling(24, min_periods=1).mean()
avg = round(avg, 3)
print('avg shape:', avg.shape)

fig = plt.figure(figsize=(16,4))

ax = fig.add_subplot(121)
ax.plot(avg)
ax.set_title('moving average')


ax = fig.add_subplot(122)
ax.plot(best_stations_df['MpKrakAlKras'])
ax.set_title('normal')
```

```python
print('lenght = ',len(best_stations_df['MpKrakAlKras'].rolling(24, min_periods=1).mean()))
print('number of na = ', best_stations_df['MpKrakAlKras'].rolling(24, min_periods=1).mean().isna().sum())
```

# plotting few days

```python
#inital values
n_days = 7
first_day = 40
period_h = 24

fig = plt.figure(figsize=(12,4))
plt.subplots_adjust(bottom=0.1, top=3, hspace = 0.5)
#plt.tight_layout(rect=(0, 0, 2, 2))

for i in range(n_days):
    ax = fig.add_subplot(n_days,2,2*i+1)
    ax.plot(avg[period_h*(i+first_day):period_h*(i+1+first_day)])
    ax.set_title('mavg day: ' + str(i+1+first_day))
    ax.set_ylim(bottom=avg.min(), top=avg.max())
    
    ax = fig.add_subplot(n_days,2,2*(i+1))
    ax.plot(best_stations_df['MpKrakAlKras'][period_h*(i+first_day):period_h*(i+1+first_day)])
    ax.set_title('normal day: ' + str(i+1+first_day))
    ax.set_ylim(bottom=best_stations_df['MpKrakAlKras'].min(), top=best_stations_df['MpKrakAlKras'].max())
```
